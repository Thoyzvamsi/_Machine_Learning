```python
import numpy as np
import pandas as pd 
import seaborn as sns
```

```python
#sns.get_dataset_names()
jew = sns.load_dataset('diamonds')
jew.head()
```

```python
sns.scatterplot(data=jew,x = "carat",y="price",hue="cut",palette="Dark2",alpha=1)
sns.set_style("whitegrid")
sns.despine(left=True)
#sns.set_context()
```

```python
sns.stripplot(data=jew,x = "carat",y="price",hue="cut",palette="Dark2")
```

```python
sns.regplot(data=jew,x = "carat",y="price")
```

```python
sns.lineplot(data = jew,x="carat" ,y="price",hue="cut")
```

```python
colomns = ["carat","depth","table","price"]
sns.heatmap(data = jew[colomns].corr() ,annot=True)
```

```python
sns.pairplot(data=jew)
```
