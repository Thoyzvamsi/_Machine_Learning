```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#Energy levels at different Days in week
data = {
    "days":["mon","tue","wed","thu","fri","sat","sun"],
    "energy":[10,20,40,60,70,100,85]
}

df = pd.DataFrame(data)
x = df["days"]
y = df["energy"]
plt.plot(x,y)
plt.grid()
salary = np.random.randint(10000,50000,size=20)
```

```python
#Univariate variable

data1 = {
    "salary": salary
}
dff = pd.DataFrame(data1)
dff.head()
plt.plot(dff["salary"],color = "green",marker="*",linestyle="--",linewidth="2")
plt.grid()
plt.show
```

```python
plt.hist(dff["salary"],bins=5)
plt.show
```

```python
plt.boxplot(dff["salary"])
plt.show
```

```python
#univariate Categorical value
dept = np.random.choice(["math","phy","che","bio"],size=20)
dff["dept"] = dept
```

```python
#piechart
coount = dff["dept"].value_counts()
coount

plt.pie(coount,labels = coount.index ,autopct = "%1.1f",explode=[0,0,0.1,0])
plt.show
```

```python
plt.bar(coount.index,coount,color=["red","green","yellow","blue"])
plt.show
```

```python
age = np.random.randint(20,45,size=20)
dff["age"] = age
```

```python
#bivariate variables
#scatter plot
plt.scatter(dff["age"],dff["salary"],color="blue")
```

```python
plt.bar(dff["age"],dff["salary"],color="red")
plt.show
```

```python
#Bivariate - categorical
mat_sal = dff.loc[dff["dept"]=="math" ,"salary"]
plt.boxplot(mat_sal,label="Maths_dept_salary")
```

```python
#bubble plot
data2 = {
    "age1" : [21,22,33,45,50],
    "exp1" : [1,3,5,10,15],
    "salary1" : [20000,25000,30000,50000,70000]
}
dfff = pd.DataFrame(data2)
plt.scatter(dfff["age1"],dfff["salary1"],s = dfff["exp1"]*20 ,edgecolors="black")
plt.show
```

```python
#multiple plots
fig , axs = plt.subplots(1,3,figsize=(15,5))
axs[0].plot(dfff["age1"],dfff["salary1"],color="blue",marker="*")
axs[0].grid()

axs[1].bar(dfff["age1"],dfff["salary1"],color="red")
axs[1].grid()

axs[2].boxplot(dfff["salary1"])
plt.savefig("savingfig.png")
```

```python
#3d plotting
ax = plt.axes(projection = "3d")
ax.scatter(dfff["age1"],dfff["exp1"],dfff["salary1"])
plt.show()
```
