```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
#sns.get_dataset_names()
df = sns.load_dataset('attention')
df.head(15)
```

```python
sns.lineplot(data = df , x = df.index,y="score",hue="attention")
```

```python
sns.barplot(data=df,x="attention",y="score")
```

```python
gpa = df["score"]
sns.barplot(data=df,x="score",y=gpa.value_counts(),palette="Set1")
```

```python
sns.boxplot(data=df,y="score",hue = "attention",palette="Dark2")
```

```python
ds = plt.axes(projection="3d")
plt.scatter(df.index,df["score"],df["solutions"])
plt.show()
```
